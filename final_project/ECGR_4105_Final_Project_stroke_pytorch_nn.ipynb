# Pytorch

```python
import torch
import torch.nn as nn
from torch.nn import functional as F
from torch.autograd import Variable
```

## Pytorch Training

```python
# credit to https://www.kaggle.com/mburakergenc/ttianic-minimal-pytorch-mlp

device = 'cuda' if torch.cuda.is_available() else 'cpu'
print('Using {} device'.format(device))

class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(11, 1331)
        self.fc2 = nn.Linear(1331, 1331)
        self.fc3 = nn.Linear(1331, 11)
        self.dropout = nn.Dropout(0.2)
        
    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.dropout(x)
        x = F.relu(self.fc2(x))
        x = self.dropout(x)
        x = self.fc3(x)
        return x

model = Net().cuda()
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
X_train
```

```python
# credit to https://www.kaggle.com/mburakergenc/ttianic-minimal-pytorch-mlp

batch_size = 500
n_epochs = 100
batch_no = len(X_train) // batch_size

train_loss = 0
train_loss_min = np.Inf

plotmx = []

for epoch in range(n_epochs):
    for i in range(batch_no):
        start = i * batch_size
        end   = start + batch_size
        x_var = Variable(torch.FloatTensor(X_train[start:end]))
        y_var = Variable(torch.LongTensor(y_train[start:end])) 
        
        optimizer.zero_grad()
        output = model(x_var)
        loss   = criterion(output,y_var)
        loss.backward()
        optimizer.step()
        
        values, labels = torch.max(output, 1)
        num_right   = np.sum(labels.data.numpy() == y_train[start:end])
        train_loss += loss.item()*batch_size

    train_loss = train_loss / len(X_train)
    train_accu = num_right / len(y_train[start:end])

    plotmx.append([epoch,train_loss,train_accu])

    if train_loss <= train_loss_min:
        print("Validation loss decreased ({:6f} ===> {:6f}). Saving the model...".format(train_loss_min,train_loss))
        torch.save(model.state_dict(), "model.pt")
        train_loss_min = train_loss
    
    if epoch % 10 == 0:
        print('')
        print("Epoch: {} \tTrain Loss: {} \tTrain Accuracy: {}".format(epoch+1, train_loss, train_accu))

print('Training Ended! ')
```

```python
def Parse(Value,i):
  return [item[i] for item in Value]
```

```python
plotEpochs = Parse(plotmx,0)
plotLoss = Parse(plotmx,1)

plt.plot(plotEpochs, plotLoss)

plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Epoch vs Train Loss')
 
plt.show()
```

```python
plotEpochs = Parse(plotmx,0)
plotAccu = Parse(plotmx,2)

plt.plot(plotEpochs, plotAccu)

plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Epoch vs Train Accuracy')
 
plt.show()
```
